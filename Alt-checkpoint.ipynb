**Convert docx to txt**

```python
from pathlib import Path  # type: ignore
import docx2txt  # type: ignore
import os

# List of DOCX files with correct relative paths
docx_files = [
    'AGGA_Africa.docx',
    'AGGA_Asia.docx',
    'AGGA_Europe.docx',
    'AGGA_NorthAmerica.docx',
    'AGGA_SouthAmerica.docx',
    'AGGA_Oceania.docx',
]

# Function to process each DOCX file
def process_docx(doc_path):
    try:
        # Check if the file exists
        if not Path(doc_path).exists():
            print(f"File not found: {doc_path}")
            return
        
        # Convert docx to plain text
        text = docx2txt.process(doc_path)
        
        # Create the corresponding TXT file path
        txt_path = Path(doc_path).with_suffix('.txt')
        
        # Save the text to a .txt file with encoding 'utf-8'
        txt_path.write_text(text, encoding='utf-8')
        
        # Output the path of the saved .txt file
        print(f"Saved: {txt_path}")
    except Exception as e:
        print(f"Error processing {doc_path}: {e}")

# Print the current working directory
print(f"Current working directory: {os.getcwd()}")

# Process each DOCX file
for docx_file in docx_files:
    # Add the working directory to the file path
    full_path = os.path.join('/Users/salehafroogh/Desktop/AGGAProject2/', docx_file)
    process_docx(full_path)
```

```python
import os
import nltk
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt

# Ensure stopwords are accessible
nltk.download('stopwords')
all_stopwords = set(stopwords.words('english')).union({
    'use', 'used', 'using', 'may', 'work', 'also', 'text', 'ways',
    'uio', "tool", "university", "provide", "relevant", "different",
    "genai", "school", "allowed"
})

# Generate a list of chart colors
chart_colors = ['#FFCCCC', '#FFE5CC', '#90EE90', '#B0C4DE', '#DDA0DD', '#AFEEEE']

# Function to process text files
def process_txt(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    # Tokenize, lowercase, and remove punctuation
    words = text.lower().split()
    return [word.strip(".,!?;:()[]{}'\"") for word in words]

# Function to plot frequency chart
def plot_frequency_chart(words, title, color):
    filtered_words = [word for word in words if word not in all_stopwords]
    counter = Counter(filtered_words)
    most_common = counter.most_common(20)

    # Print the most common words
    print(f"Top 20 words in {title}:")
    for word, frequency in most_common:
        print(f"{word}: {frequency}")
    
    # Prepare data for plotting
    if most_common:
        words, frequencies = zip(*most_common)
        colors = chart_colors[:len(words)]
        plt.figure(figsize=(10, 6))
        plt.bar(words, frequencies, color=colors)
        plt.xlabel('Words')
        plt.ylabel('Frequencies')
        plt.title(f'Top 20 Keywords in {title}')
        plt.xticks(rotation=45)
        plt.show()
    else:
        print(f"No significant words found in {title}.")

# List of text files to process
txt_files = [
    "AGGA_Africa.txt",
    "AGGA_Asia.txt",
    "AGGA_Europe.txt",
    "AGGA_NorthAmerica.txt",
    "AGGA_Oceania.txt",
    "AGGA_SouthAmerica.txt"
]

# Process each file and generate a plot
for i, txt_file in enumerate(txt_files):
    file_path = os.path.join('/Users/salehafroogh/Desktop/AGGAProject2/', txt_file)  # Adjust to your directory
    if os.path.exists(file_path):
        words = process_txt(file_path)
        plot_frequency_chart(words, txt_file.replace('.txt', ''), chart_colors[i % len(chart_colors)])
    else:
        print(f"File not found: {file_path}")
```
